```python
# -*- coding: utf-8 -*-
# <nbformat>4.1</nbformat>

# <codecell>

from IPython.display import YouTubeVideo, display, HTML

# Adicionando um vídeo usando a função do IPython.display
# Basta copiar a parte final do URL do YouTube, que é o ID do vídeo.
# Por exemplo, para 'https://www.youtube.com/watch?v=dQw4w9WgXcQ', o ID é 'dQw4w9WgXcQ'.

print("--- Usando o método IPython.display.YouTubeVideo ---")
print("Exemplo 1: Vídeo padrão")
# Substitua o ID abaixo pelo ID do vídeo que você deseja exibir.
video_id_1 = "7zu6I8JUvBM"
print(f"Tentando exibir o vídeo com o ID: {video_id_1}")
display(YouTubeVideo(video_id_1))
print("\nSe o vídeo não for exibido, verifique sua conexão com a internet e se o vídeo está disponível publicamente.")


# Você também pode controlar o tamanho da exibição
print("\nExemplo 2: Vídeo com tamanho personalizado (largura=400, altura=225)")
video_id_2 = "E0N108g5S0U"
display(YouTubeVideo(video_id_2, width=400, height=225))

# <codecell>

# Adicionando um vídeo usando um iframe HTML
# Este método é útil para ter controle mais granular sobre a formatação.

print("--- Usando um iframe HTML ---")
print("Exemplo 3: Vídeo com iframe HTML e tamanho personalizado")

video_id_3 = "tgbNymZ7vqY"
html_code = f"""
<iframe width="560" height="315"
src="https://www.youtube.com/embed/{video_id_3}"
frameborder="0"
allow="accelerometer; autoplay; encrypted-media; gyroscope; picture-in-picture"
allowfullscreen>
</iframe>
"""
display(HTML(html_code))

# <codecell>

# Exibindo um iframe HTML com estilo centralizado

print("\nExemplo 4: Vídeo centralizado")

video_id_4 = "dQw4w9WgXcQ"
html_code_centered = f"""
<div style="text-align: center;">
    <iframe width="560" height="315"
    src="https://www.youtube.com/embed/{video_id_4}"
    frameborder="0"
    allow="accelerometer; autoplay; encrypted-media; gyroscope; picture-in-picture"
    allowfullscreen>
    </iframe>
</div>
"""
display(HTML(html_code_centered))
```
