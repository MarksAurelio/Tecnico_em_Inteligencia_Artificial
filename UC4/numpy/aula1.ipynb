Leitura do Histograma

- Eixo X (Horizontal - "Valor"): Representa os valores da coluna 1 do dataset Iris. O eixo está dividido em intervalos, chamados de bins (compartimentos ou classes).

    - Exemplo dos bins: Os valores vão de 2.0 a 4.5, com divisões claras em 2.0, 2.5, 3.0, 3.5, 4.0, e 4.5.

- Eixo Y (Vertical - "Frequência"): Representa a contagem ou o número de ocorrências (frequência) de valores que caem dentro de cada intervalo (bin) do eixo X.

🔍 Interpretação dos Dados

- Analisando a altura das barras, podemos tirar as seguintes conclusões:

- Maior Frequência (Moda): A barra mais alta está centrada ligeiramente acima do valor 3.0.

    - Isso significa que a maior concentração de valores na coluna 1 do dataset Iris está no intervalo entre aproximadamente 2.9 e 3.1 (o bin correspondente a 3.0 tem uma frequência de 38, o valor máximo no Y).

- Segunda Maior Frequência: O segundo pico está no intervalo entre 3.0 e 3.5. A barra centrada em 3.25 tem uma frequência de cerca de 31.

- Distribuição dos Dados:

    - A distribuição não é perfeitamente simétrica. Ela parece ter um leve viés (assimetria) para a direita (a cauda estende-se um pouco mais para a direita), mas é mais ou menos em forma de sino (distribuição normal) em sua parte principal.

    - Valores Baixos: Os valores mais baixos, no intervalo de 2.0 a 2.25, têm uma frequência baixa (cerca de 4).

    - Valores Altos: Os valores mais altos (acima de 4.0) também têm uma frequência muito baixa (cerca de 2).

- Conclusão: A maioria dos dados (a maior parte da coluna 1) está concentrada na faixa de 2.5 a 3.5, com um pico muito forte por volta de 3.0.
- Em resumo, o histograma mostra que a característica representada pela coluna 1 no dataset Iris varia principalmente entre 2.0 e 4.5, sendo 3.0 o valor mais comum.

Usar o MatplotLib com o ret (retorno financeiro)
- O gráfico deve ficar semelhante ao modelo:

![Alt text](grafico_exemplo.png)

- Formula implementada no numpy
- Usar matplotlib
- dica (fig)

```python
import matplotlib.pyplot as plt

# 1. Criação da Série Histórica de Preços
# Vamos simular 20 pontos de preço (para ter 19 retornos)
np.random.seed(42)  # Para resultados reprodutíveis
precos_simulados = 100 + np.cumsum(np.random.normal(0, 5, 20))
# Convertendo para uma Series do Pandas para facilitar o cálculo
precos = pd.Series(precos_simulados)

# 2. Implementação da Fórmula de Retorno (ret) usando NumPy/Pandas

# O método mais eficiente em Pandas/NumPy é usar .pct_change()

# preco(i) é o preço atual, e preco(i-1) é o preço anterior (lag=1)
retornos = precos.pct_change() 

# pct_change(n) calcula: (Valor Atual - Valor Anterior n períodos) / Valor Anterior n períodos
# O primeiro valor (retornos[0]) é NaN porque não há preço anterior (i-1) para o primeiro preço (i=0).

# Para replicar o gráfico, normalmente queremos que os retornos variem
# de -0.06 a 0.06. Vamos ajustar os retornos para essa escala.
retornos_ajustados = retornos * 20 # Multiplicamos por um fator para simular a variação do gráfico de exemplo
retornos_ajustados[0] = 0 # Definindo o primeiro NaN como 0 para iniciar o gráfico

# 3. Plotando o Gráfico com Matplotlib

# Dica (fig): Cria uma figura e um conjunto de eixos
fig, ax = plt.subplots(figsize=(10, 6)) 

# Plotagem da linha
# Usamos o índice (tempo) no eixo X e os retornos no eixo Y
ax.plot(retornos_ajustados, color='dimgray', linewidth=1.5)

# Configurações do Gráfico
ax.set_title("Retorno Financeiro (Ret)", fontsize=14, fontweight='bold')
ax.set_xlabel("tempo", fontsize=12)
ax.set_ylabel("retorno", fontsize=12, rotation=90) 
ax.grid(axis='y', linestyle='--', alpha=0.7) # Adiciona linhas de grade horizontais

# Ajustando o formato do eixo Y para ser mais limpo (como no gráfico de exemplo)
from matplotlib.ticker import FormatStrFormatter
ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))

# Exibir o gráfico
plt.show()
```
