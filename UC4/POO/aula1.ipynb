* Lista do Python x Array do Numpy

Porque o Numpy Array e mais rápido que uma lista Python

- O Numpy permite a vetorização

- Essa é a dúvida mais fundamental (e importante) para quem está migrando do Python "puro" para Data Science.

- A diferença principal resume-se a: Listas são flexíveis, Arrays são eficientes.

- Imagine a seguinte analogia:

* Lista Python: É uma sacola de compras onde você joga qualquer coisa dentro (uma maçã, uma chave de fenda, uma carta). É fácil de usar, mas bagunçada para processar.

* Array NumPy: É uma caixa de ovos. Ela só aceita um tipo de item (ovos), tem tamanho fixo, mas é extremamente organizada e fácil de empilhar e transportar.
