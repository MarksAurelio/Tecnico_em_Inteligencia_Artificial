```python
mascara_nulos = df_total.isnull()
print(mascara_nulos)
```

```python
# Apenas na coluna 'Year', preenche com 0
df_total['Year'] = df_total['Year'].fillna(0)

# Apenas na coluna 'Title', preenche com 'Desconhecido'
df_total['Title'] = df_total['Title'].fillna('Desconhecido')

print(df_total)
```

```python
df_total.describe()
```

```python
df_pivotado = df_total.pivot(index='Rank', columns='Year', values='Rating')

print("\n--- DataFrame Pivotado (Wide) ---")
print(df_pivotado)
```

```python
# Removendo colunas que não serão usadas na análise
df_limpo = df_total.drop(columns=['IMDb URL', 'Image URL'])

print(df_limpo.head())
```

```python
# 2. Corrigir a conversão usando lambda com uma verificação condicional (if/else)
df_total['Duration_Minutes'] = df_total['Duration'].apply(
    lambda x: (
        # Se 'h' estiver presente (formato 'Xh Ym')
        int(x.split('h')[0].strip()) * 60 + int(x.split('h')[1].replace('m', '').strip())
        if 'h' in x
        # Senão (formato 'Ym')
        else int(x.replace('m', '').strip())
    )
)

print("As 5 primeiras e a linha do 'Sherlock Jr.' convertidas:")
print(df_total.iloc[[0, 205]][['Title', 'Duration', 'Duration_Minutes']])
```

```python
media_rating = df_total['Rating'].mean()
print(f"A Média de Avaliação (Rating) dos {len(df_total)} filmes é: {media_rating:.2f}")
```

```python
# A mediana da coluna 'Rating'
mediana = df_total['Rating'].median()
print(mediana)
```

```python
# A moda da coluna 'Rating' 
mode = df_total.Rating.mode()
print(mode)
```
